<h3>4.1 Clustering Neighborhoods of Toronto:</h3>


First step in K-means clustering is to identify best K value in a given dataset - which means the number of clusters we can optimaly work with. <br>To do so we are going to use the elbow method on the Toronto dataset with chinese restaurant percentage

```python
#make dataframe
df_toronto_part = pd.DataFrame(toronto_grouped[['Neighbourhood', 'Chinese Restaurant']])

#df_toronto_part.head()

# Chinese Restaurant
df_toronto_part.sort_values(by=['Chinese Restaurant'], inplace=True, ascending=[False])
df_toronto_part
```

```python
from sklearn.cluster import KMeans

df_toronto_part_clustering = df_toronto_part.drop('Neighbourhood', 1)

error_cost = []

for i in range(3,11):
    KM = KMeans(n_clusters = i, max_iter = 100)
    try:
        KM.fit(df_toronto_part_clustering)
    except ValueError:
        print("error on line",i)
    
    
    
    
    #calculate squared error for the clustered points
    error_cost.append(KM.inertia_/100)

#plot the K values aganist the squared error cost
plt.plot(range(3,11), error_cost, color='r', linewidth='3')
plt.xlabel('K values')
plt.ylabel('Squared Error (Cost)')
plt.grid(color='white', linestyle='-', linewidth=2)
plt.show()
```
